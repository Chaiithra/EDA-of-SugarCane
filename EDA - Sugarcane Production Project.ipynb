```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("List of Countries by Sugarcane Production.csv")
```

```python
df.shape
```

## Renaming a column

```python
df = df.rename(columns={df.columns[0]:"index"})
```

## Dropping a column

```python
df = df.drop(columns='index')
```

```python
df.shape
```

## Understanding Data and it's Preprocessing

```python
df.info()
```

## Changing their data type 

```python
df['Country']= df['Country'].astype(str)
```

```python
df['Continent']= df['Continent'].astype(str)
```

```python
df["Production (Tons)"]= df["Production (Tons)"].str.replace(".","")
df["Production (Tons)"]= df["Production (Tons)"].str.replace(",","")

df["Production per Person (Kg)"] = df["Production per Person (Kg)"].str.replace(".","")
df["Production per Person (Kg)"] = df["Production per Person (Kg)"].str.replace(",","")

df["Yield (Kg / Hectare)"]= df["Yield (Kg / Hectare)"].str.replace(",","")
df["Yield (Kg / Hectare)"]= df["Yield (Kg / Hectare)"].str.replace(".","")

df["Acreage (Hectare)"] = df["Acreage (Hectare)"].str.replace(",","")
df["Acreage (Hectare)"] = df["Acreage (Hectare)"].str.replace(".","")
```

```python
df['Production (Tons)']= df['Production (Tons)'].astype(int)
```

```python
df['Production per Person (Kg)']= df['Production per Person (Kg)'].astype(int)
```

```python
df['Acreage (Hectare)']= df['Acreage (Hectare)'].astype(float)
```

```python
df['Yield (Kg / Hectare)']= df['Yield (Kg / Hectare)'].astype(float)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.dropna(inplace = True)
```

```python
df.shape
```

## Checking Outliers

```python
plt.figure(figsize = (10,8))

plt.subplot(2,2,1)
sns.boxplot(df["Production (Tons)"])
plt.title("Production (Tons)")

plt.subplot(2,2,2)
sns.boxplot(df["Production per Person (Kg)"])
plt.title("Production per Person (Kg)")

plt.subplot(2,2,3)
sns.boxplot(df["Acreage (Hectare)"])
plt.title("Acreage (Hectare)")

plt.subplot(2,2,4)
sns.boxplot(df["Yield (Kg / Hectare)"])
plt.title("Yield (Kg / Hectare)")

plt.show()
```

we have outliers in the data but outliers are required here as it shows the countries which has maximum production. And then we can see what are the reasons for outliers.

# Analysis for Continent

```python
df4 = df.copy()
```

```python
df4 = df.groupby("Continent").sum()
```

```python
df4["Number of Countries"] = df.groupby("Continent").count()["Country"]
```

```python
df4
```

# Which continent produces maximum sugarcane?

```python
df4["Production (Tons)"].plot(kind='pie', autopct = "%2f")
```

```python
df4["Production (Tons)"].sort_values(ascending = False).plot(kind = 'bar')
```

```python
df4.corr()
```

# Do number of countries in a Continent effects production of sugarcane?

```python
sns.jointplot(x=df4["Number of Countries"],y=df4["Production (Tons)"])
```

# Does number of Countries in a Continent effect the Yield?

```python
continent_names = df4.index.to_list()
sns.lineplot(data = df4, x = "Number of Countries", y = "Yield (Kg / Hectare)")
plt.xticks(df4["Number of Countries"],continent_names, rotation = 90)
plt.show()
```
